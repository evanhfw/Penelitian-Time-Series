```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# from XGBoost import XGBoostRegressor
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
data = pd.read_csv('data/data.csv')
data
```

```python
X = data.drop(['Consumption'], axis=1)
y = data[['Consumption']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=31*24, shuffle=False)
```

```python
model = XGBRegressor(random_state=42, n_estimators=100000, early_stopping_rounds=1000, eval_set=[(X_test, y_test)], eval_metric='mape')
model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

```python
pd.Series(model.predict(X_test)).to_excel('prediction.xlsx')
```
